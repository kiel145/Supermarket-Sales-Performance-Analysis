# Import Library

```python
# Import Library
from great_expectations.data_context import FileDataContext
```

# Instantiate Data Context

```python
# Create Data Context

context = FileDataContext.create(project_root_dir='./')
```

# Connect to A Datasource

```python
# Jalankan ini untuk membersihkan memori GX
context.delete_datasource('csv-milestone-data')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-milestone-data'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'Milestone3'
path_to_data = 'P2M3_kevin_hibatul_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

# Create an Expectation Suite

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-supermart-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

## Expectations

**Expectation 1 — to be unique**

```python
validator.expect_column_values_to_be_unique(column="order_id")
```

**Expectation 2 — to be between min & max**

```python
validator.expect_column_values_to_be_between(
    column='discount',
    min_value=0,
    max_value=1
)
```

**Expectation 3 — to be in set**

```python
validator.expect_column_values_to_be_in_set(
    column='category',
    value_set=[
        'Food Grains',
        'Fruits & Veggies',
        'Bakery',
        'Beverages',
        'Snacks',
        'Eggs, Meat & Fish',
        'Oil & Masala'
    ]
)
```

**Expectation 4 — to be in type list**

```python
validator.expect_column_values_to_be_in_type_list(
    column='sales',
    type_list=['int', 'float', 'int64', 'float64']
)
```

**Expectation 5 — values to not be null**

```python
validator.expect_column_values_to_not_be_null(
    column='order_date'
)
```

**Expectation 6 — mean to be between**

```python
validator.expect_column_mean_to_be_between(
    column='sales',
    min_value=10,
    max_value=5000
)
```

**Expectation 7 — values to match regex**

```python
validator.expect_column_values_to_not_be_null('order_id')
```

## Validasi Akhir & Save

```python
results = validator.validate()
results['success']
```

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```

## Checkpoint

```python
# Create a checkpoint

checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint

checkpoint_result = checkpoint_1.run()
```
